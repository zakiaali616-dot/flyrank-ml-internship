## Decision
Decide whether a credit card transaction should be approved or flagged as potentially fraudulent.

## Action
If a transaction is predicted as fraudulent, block it or send it for manual review. Otherwise, approve the transaction.

## Cost of a Wrong Decision
A false negative may allow fraud and cause financial loss. A false positive may block a legitimate customer and reduce customer trust.

## Why Data Can Help
Historical transaction data such as transaction amount, time, merchant information, location, and customer behavior can help identify fraudulent transactions.
